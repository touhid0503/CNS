#Transposition Cipher

```python
def enc(pt, w=4):
  length=len(pt)
  ct=""
  for i in range(w):
    for j in range(i,length,w):
      ct+=pt[j]
  return ct
```

```python
def dec(ct, w=4):
  length = len(ct)
  pt_list = [''] * length
  ptr = 0

  for i in range(w):
    for j in range(i, length, w):
      if ptr < length:
        pt_list[j] = ct[ptr]
        ptr += 1
      else:
        break

  return "".join(pt_list)
```

```python
print(enc("DEPARTMENT OF COMPUTER SCIENCE AND ENGINEERING VARENDRA UNIVERSITY BANGLADESH",4))
```

```python
print(dec("DRNFMECCNNENADUETAAHETT PRIEDGEGRRNRYNDPM CU E  IR EAIS GEAEOOTSNAENIVN VIBLS",4))
```
