#Plaintext to Ciphertext

Encrypt function

```python
def encrypt(plaintext):
    ciphertext=""
    for i in plaintext:
        if i.isalpha():
            i=chr(ord(i)+3)
            if i.isalpha()!=True:
                i=chr(ord(i)-26)
        ciphertext+=i

    return ciphertext
```

Decrypt function

```python
def decrypt(plaintext):
    ciphertext=""
    for i in plaintext:
        if i.isalpha():
            i=chr(ord(i)-3)
            if i.isalpha()!=True:
                i=chr(ord(i)+26)
        ciphertext+=i

    return ciphertext
```

calling functions

```python
plaintext="TOUHID123"
print(encrypt(plaintext))
```

```python
plaintext="TOUHID123"
print(decrypt(plaintext))
```
